```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
```

##Load Data

```python
df = pd.read_csv("data/final_dataset.csv")

df.head()
```

```python
df.describe()
```

## Correlation Between AI Capability and Risk Score

```python
corr, p_value = pearsonr(df["ai_capability"], df["risk_score"])

print("Correlation:", corr)
print("P-value:", p_value)
```

## Correlation Matrix

```python
correlation_matrix = df.corr(numeric_only=True)

correlation_matrix
```
