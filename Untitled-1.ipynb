```python
import matplotlib_inline as plt
import sklearn
import numpy as np
import pandas as pd
```

# Load datasets

```python

co2 = pd.read_csv("data/annual-co2-emissions-per-country.csv")

temp = pd.read_csv("data/GLB.Ts+dSST.csv", skiprows=1)

conflict = pd.read_csv("data/BattleDeaths_v25_1_conf.csv")

ai = pd.read_csv("data/annual-scholarly-publications-on-artificial-intelligence.csv")

print("All datasets loaded successfully")

print(co2.head())
print(temp.head())
print(conflict.head())
print(ai.head())
```

## Clean CO₂ Data



```python
co2 = co2[co2["Entity"] == "World"]

co2 = co2[["Year", "Annual CO₂ emissions"]]

co2.rename(columns={
    "Year": "year",
    "Annual CO₂ emissions": "co2"
}, inplace=True)

co2 = co2.dropna()

co2.head()
```

##  Clean Temperature Data




```python
temp.columns = temp.columns.str.strip()

temp.rename(columns={temp.columns[0]: "year"}, inplace=True)

temp["year"] = temp["year"].astype(int)

monthly_cols = temp.columns[1:13]

temp[monthly_cols] = temp[monthly_cols].apply(pd.to_numeric, errors="coerce")

temp["temp_anomaly"] = temp[monthly_cols].mean(axis=1)

temp = temp[["year", "temp_anomaly"]]

temp = temp.dropna()

temp.head()
```

## Clean Conflict Data



```python
conflict = conflict.groupby("year")["bd_best"].sum().reset_index()

conflict.rename(columns={"bd_best": "conflict_deaths"}, inplace=True)

conflict.head()
```

## Clean AI Data



```python
ai.columns = ai.columns.str.strip()

print(ai.columns)
```
